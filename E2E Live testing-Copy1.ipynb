```python
import pandas as pd
from fyers_apiv3.FyersWebsocket import data_ws
import json
from datetime import datetime

# Initialize an empty DataFrame
columns = ["symbol", "ltp", "ch", "chp", "open_price", "high_price", "low_price", "prev_close_price", "exch_feed_time"]
df = pd.DataFrame(columns=columns)

def onmessage(message):
    """
    Callback function to handle incoming messages from the FyersDataSocket WebSocket.

    Parameters:
        message (dict): The received message from the WebSocket.
    """
    global df
    
    try:
        # Ensure message is a dictionary
        if isinstance(message, str):
            message = json.loads(message)
        
        # Extract relevant fields
        data = {
            "symbol": message.get("symbol"),
            "ltp": float(message.get("ltp", 0)),
            "ch": float(message.get("ch", 0)),
            "chp": float(message.get("chp", 0)),
            "open_price": float(message.get("open_price", 0)),
            "high_price": float(message.get("high_price", 0)),
            "low_price": float(message.get("low_price", 0)),
            "prev_close_price": float(message.get("prev_close_price", 0)),
            "exch_feed_time": datetime.fromtimestamp(int(message.get("exch_feed_time", 0)))
        }
        
        # Append new data to DataFrame
        df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)

        # Print DataFrame in real-time
        print(df.tail())

    except Exception as e:
        print("Error processing message:", e)


def onerror(message):
    """Handles WebSocket errors."""
    print("Error:", message)


def onclose(message):
    """Handles WebSocket disconnection events."""
    print("Connection closed:", message)


def onopen():
    """Subscribes to market index data upon WebSocket connection."""
    data_type = "SymbolUpdate"
    symbols = ["NSE:NIFTY50-INDEX", "NSE:NIFTYBANK-INDEX"]
    fyers.subscribe(symbols=symbols, data_type=data_type)
    fyers.keep_running()


# Replace with your actual access token
access_token = "XC4XXXXXXM-100:eXXXXXXXXXXXXfZNSBoLo"

# Initialize WebSocket connection
fyers = data_ws.FyersDataSocket(
    access_token=access_token,
    log_path="",
    litemode=False,
    write_to_file=False,
    reconnect=True,
    on_connect=onopen,
    on_close=onclose,
    on_error=onerror,
    on_message=onmessage
)

# Establish WebSocket connection
fyers.connect()
```
